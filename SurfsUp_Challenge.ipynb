```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import pandas as pd
import datetime as dt
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
from matplotlib.dates import DateFormatter, DayLocator
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract
## sqlalchemy.sql.expression.extract(field, expr, **kwargs)
#  3. Convert the June temperatures to a list.
june_temps = []
# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temps = session.query(Measurement.date, Measurement.tobs).\
filter(extract('month', Measurement.date) == 6).all()
print(june_temps)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_temps, columns=['date', 'temps'])
june_df['date'] = pd.to_datetime(june_df['date'])
june_df.set_index(june_df['date'], inplace=True)
# Sort the dataframe by date
june_df = june_df.sort_index()
```

```python
print(june_df.to_string(index=False))
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_df.describe()
```

```python
### Plotted June Data.
#june_df.

x= june_df['date'].dt.year
y= june_df['temps']
fig, ax = plt.subplots(figsize=(10,8))
ax.bar(x, y, color='green', label='Temp/time')
ax.axhline(y=74.94, linewidth=2, color='r', label='Avg_temp')
ax.xaxis.set_major_formatter(DateFormatter('%Y/%m'))
ax.set_xlabel('Date')
ax.set_ylabel('Temp (F)')
ax.set_title('June Temps')
ax.grid(True)
ax.legend(loc='upper right');
plt.show()
```
